```python
#(c) Samia Mohinta 2024

import nibabel
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.morphology import medial_axis

from skimage.morphology import skeletonize

image = nibabel.load('/Users/jamesruffle/Desktop/UCSF-PDGM-0005_enh.nii.gz')
print(image.shape)
img = image.get_fdata()
print(np.unique(img))
print(np.argwhere(img == 1))
plt.imshow(img[..., 92], cmap='gray')

skeleton = skeletonize(img)
plt.imshow(skeleton[..., 92], cmap='gray')

allpixels = np.count_nonzero(img)
skeletonpixels = np.count_nonzero(skeleton)
print(allpixels/skeletonpixels)
plt.show()
```
